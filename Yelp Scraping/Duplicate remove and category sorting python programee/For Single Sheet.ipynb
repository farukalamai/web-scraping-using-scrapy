# Remove duplicates values from a single Sheet

```python
import pandas as pd

# From input file
df = pd.read_csv("input/austin.csv")

# Removing merged cell
df.fillna(value='', inplace=True)

# shorting and Removing duplicate

df.drop_duplicates(inplace=True)


df.sort_values("Yelp URL", inplace=True)
df.drop_duplicates(subset="Yelp URL",
                     keep="first", inplace=True)

df.head()


# Removing others state data

# df.to_excel("New_Jersey.xlsx", index=False)
```

```python
# Removing others state data
current_state_data = df[df['State'] == 'TX']


current_state_data
```

```python
current_state_data.to_csv("Austin Resturant Data.csv", index=False)
```
