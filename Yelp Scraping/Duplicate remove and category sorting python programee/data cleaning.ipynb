# Remove duplicate from multiple sheet

```python
import pandas as pd

# From input file
df1 = pd.read_csv("input/austin.csv")
df2 = pd.read_csv("input/texasone.csv")
df3 = pd.read_csv("input/texastwo.csv")



# Adding csv file
all_data = pd.concat([df1,df2,df3])

all_data
```

```python
# shorting and Removing duplicate

# all_data.drop_duplicates(inplace=True)
# all_data['State'].str.replace(' ', '')


all_data.sort_values("Yelp URL", inplace=True)
all_data.drop_duplicates(subset="Yelp URL",
                     keep="first", inplace=True)

# all_data.to_csv("Dermatolgistis and Plastic Surgeon.csv", index=False)
```

```python
# Removing others state data
current_state_data = all_data[all_data['State'] == 'TX']


current_state_data.to_csv("Austin and Houston.csv", index=False)



# Category sorting data [to remove others categories data]

# Final output
```
